Analyzing the top 5 Actors on Netflix

```python
#Replace missing cast values
df['cast'] = df["cast"].fillna("no cast specified")
```

```python
#Create empty dataframe
cast_df = pd.DataFrame()
```

```python
#Split multiple actors
cast_df = df['cast'].str.split(',', expand=True).stack()
```

```python
#Convert Series → DataFrame
cast_df = cast_df.to_frame()
```

```python
#Rename column
cast_df.columns = ['actor']
```

```python
#remove unwanted rows
actors = cast_df.groupby(['actor']).size().reset_index(name='Total Count')
```

```python
#Remove fake values
actors = actors[actors.actor != 'no cast specified']
```

```python
#Sort highest → lowest
actors = actors.sort_values(by=['Total Count'], ascending=False)
```

```python
#Get top 5 actors
top5Actors = actors.head()
```

```python
#Sort again for chart appearance
top5Actors = top5Actors.sort_values(by=['Total Count'])
```

```python
#Create bar chart
barChart2 = px.bar(
    top5Actors,
    x='Total Count',
    y='actor',
    title='Top 5 Actors on Netflix'
)
```

```python
#show chart
barChart2.show()
```

Sentiment Analysis of Netflix Content

```python
#Select needed columns
df3 = df[['release_year', 'description']]
```

```python
#Rename columns
df3 = df3.rename(columns={
    'release_year':'Release Year',
    'description':'Description'
})
```

```python
#Loop through each row
for index, row in df3.iterrows():
  d=row['Description']
  testimonial = TextBlob(d)
  p = testimonial.sentiment.polarity
  if p==0:
    sent = 'Neutral'
  elif p>0:
    sent = 'Positive'
  else:
    sent = 'Negative'
  df3.loc[[index, 2], 'Sentiment']=sent
```

```python
#Save sentiment into dataframe
df3.loc[index, 'Sentiment'] = sent
```

```python
#Count sentiments per year
df3 = df3.groupby(
    ['Release Year', 'Sentiment']
).size().reset_index(name='Total Count')
```

```python
#Filter years after 2005
df3 = df3[df3['Release Year'] > 2005]
```

```python
#Create visualization
barGraph = px.bar(
    df3,
    x="Release Year",
    y="Total Count",
    color="Sentiment",
    title="Sentiment Analysis of Content on Netflix"
)
barGraph.show()
```
